```python
# Input Output Creation
ip = adv.drop('Sales',axis=1)
op = adv.Sales
```

```python
ip.head()
```

```python
op.head()
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(ip,op,test_size=0.2,random_state=3)
```

```python
x_train
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
# Standard Scaler
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()

x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
print(x_train)
print(x_test)
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
#Predication
pred = lr.predict(x_test)
pred
```

```python
y_test
```

```python
#Mean square error
#Accuracy
from sklearn.metrics import mean_squared_error , r2_score

mse = mean_squared_error(pred,y_test)
r2 = r2_score(pred,y_test)

print('MSE : ',mse) #mse --> error percentage
print('R2 Score : ',r2) #r2 score --> accuracy percentage
```

```python
#lmplot - Linear model plot
df = pd.DataFrame({'Y_Test':list(y_test),'Prediction':pred}) #here we converted the y_test to list to not to get the random index number
df
```

```python
sns.lmplot(x = 'Y_Test',y = 'Prediction',data = df)
plt.show()
```
