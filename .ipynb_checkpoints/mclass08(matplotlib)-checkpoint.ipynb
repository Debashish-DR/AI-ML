```python
#MATPLOTLIB
#For graphically data visualisation

import matplotlib.pyplot as plt
```

```python
#Line graph
x = [2,3,4,5,6,9,2]
y = [9,8,7,6,5,4,3]
plt.plot(x,y,color='r') # change color to red ----we can use ':'->for dotted line and many more and 'o'for rounded co-ordinates and 's' for square and many more
plt.title('Line Plot')
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.grid() #for graph lines
plt.show()
```

```python
#Bar Plot
x = [2,4,6,8,10]
y = [1,3,5,7,9]
plt.bar(x,y,color='y')
plt.title('Bar Plot')
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.show()
```

```python
#multiple bar plot(single graph) can be overlapped!!!

x = [2,4,6,8,10]
y = [1,3,5,7,9]
plt.bar(x,y,color='r',label = 'Example1')

x1 = [4,7,2,9,8]
y1 = [9,2,6,4,1]
plt.bar(x1,y1,color='orange',label = 'Example2')



plt.title('Bar Plot')
plt.legend() # for showing examples...
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.show()
```

```python
#multiple bar plot (separate graphs)
x = [2,4,6,8,10]
y = [1,3,5,7,9]
plt.bar(x,y,color='r',label = 'Example1')
plt.show()

x1 = [4,7,2,9,8]
y1 = [9,2,6,4,1]
plt.bar(x1,y1,color='orange',label = 'Example2')
plt.show()

plt.title('Bar Plot')
plt.legend() # for showing examples...
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.show()
```

```python
x = [2,4,6,8,10]
y = [1,3,5,7,9]
plt.bar(x,y,color='r',label = 'Example1')

x = [2,4,6,8,10]
y = [1,3,5,7,9]
plt.plot(x1,y1,color='b',label = 'Example2')

plt.title('Bar Plot')
plt.legend() # for showing examples...
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.show()
```

```python
#Histogram

ages = [20,10,15,2,8,17,30,25,18,35,45,42,40]
bins = [10,20,30,40,50]

plt.hist(ages , bins , histtype = 'bar',rwidth = 0.8) #rwidth-->bar width ,, 
plt.title('Histogram')
plt.show()
```

```python
#Scatter plot
x = [3,4,5,7,2,10,7,8]
y = [2,6,4,2,4,7,10,9]
plt.scatter(x,y,s = 500,color='r',marker = '*') # s is for size ,,, marker='o'(default) is for changing the icon syle
plt.title('Scatter Plot')
plt.show()
```

```python
#Flower plot

x = [3,4,5,7,2,10,7,8]
y = [2,6,4,2,4,7,10,9]

plt.bar(x,y,color = 'green',width = 0.05)
plt.scatter(x,y,s = 500,color='yellow',marker = '*',edgecolors='orange',linewidths = 10)
plt.scatter(x,y,color='brown')

plt.title('Flower Plot')
plt.show()
```

```python
# Pie chart
slices = [10 , 5 , 8 , 3 , 4]
activities = ['Phone' , 'Eating' , 'Sleeping' , 'Studying' , 'Playing']
cols = ['b','gold','r','m','g']

plt.pie(slices , labels = activities,colors = cols , startangle = 90)
plt.title('Pie Chart')
plt.show()
```

```python
slices = [10 , 5 , 8 , 3 , 4]
activities = ['Phone' , 'Eating' , 'Sleeping' , 'Studying' , 'Playing']
cols = ['b','gold','r','m','g']

plt.pie(slices , labels = activities,colors = cols , startangle = 90 , shadow = True , explode=(0,0,0.1,0,0)) #shadow --> for shadow ,,, explode--> for separating
plt.title('Pie Chart')
plt.show()
```

```python
slices = [10 , 5 , 8 , 3 , 4]
activities = ['Phone' , 'Eating' , 'Sleeping' , 'Studying' , 'Playing']
cols = ['b','gold','r','m','g']

plt.pie(slices , labels = activities,colors = cols , startangle = 90 , shadow = True , explode=(0,0,0.1,0,0),autopct = '%1.1f%%') #autopct --> for percentage
plt.title('Pie Chart')
plt.show()
```

```python
# SEABORN Seaborn
# Statistical Data Visualization Technique
# Nature of Data --> 1.Contineous(contineous numbers) 2.Categorical(eg.male/female , yes/no , true/false , 0/1 --> it can be String or integer)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv(r"D:\jupyter_PY\CTTC-AIML\iris\iris.data", header = None)
iris
```

```python
iris.columns = ['SL','SW','PL','PW','FlowerType']
iris
```

```python
#Data Cleaning
iris.isnull().sum()
```

```python
for i in iris.columns:
    print(i,':','\n',iris[i].unique(),'\n')
```

```python
import warnings
warnings.filterwarnings('ignore') # for ignore warnings
```

```python
# Distribution Plot(distplot) --> (it uses continous and catagorical data) & (it is the combination of -- Histogram plot(hsit plot or dis plot) and Kde plot(kernal density estimate)) 

sns.distplot(iris.SL)
plt.show()
```

```python
sns.kdeplot(iris.SL,fill=True,color='purple')
plt.show()
```

```python
sns.histplot(iris.SL)
plt.show()
```

```python
sns.histplot(iris.SL,fill=False,color='g',hatch='/')
plt.show()
```
