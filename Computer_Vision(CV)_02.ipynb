```python
#video feed
vdo = r"https://youtu.be/XQr2KmBQ7A4?si=oG2hlRINCMf8SIsW"
cv2.namedWindow(vdo)
cap = cv2.VideoCapture(vdo)
if cap.isOpened():
    ret,frame = cap.read()
    print(ret)
    print(frame)
else:
    ret = False
while ret:
    ret , frame = cap.read()
    if not ret:
        break
    cv2.imshow(vdo,frame)
    if cv2.waitKey(30)==27:
        break
cv2.destroyAllWindows()
cap.release()
```

```python
#Face Detection useing Haarcascade
face = r"D:\jupyter_PY\CTTC-AIML\haarcascade_frontalface_default.xml"
model = cv2.CascadeClassifier(face)

img = cv2.imread(r"D:\jupyter_PY\CTTC-AIML\Virat-Kohli.jpg")

facepoints = model.detectMultiScale(img) #this finds the co-ordinate points automatically

for x,y,w,h in facepoints:
    cv2.rectangle(img,(x,y),(x+w,y+h) , (0,0,255),4)

plt.imshow(img[::,::,::-1])
plt.show()
```

```python
#Face Detection using web cam
face = r"D:\jupyter_PY\CTTC-AIML\haarcascade_frontalface_default.xml"
cap = cv2.VideoCapture(0)

ret,frame = cap.read()

while ret:
    ret,frame = cap.read()
    print(ret)
    print(frame)

    if not ret:
        break
    model = cv2.CascadeClassifier(face)
    facepoints = model.detectMultiScale(frame)

    for x,y,w,h in facepoints:
        cv2.rectangle(frame,(x,y),(x+y,y+h),(0,0,255),4)

    cv2.imshow('My Face',frame)
    if cv2.waitKey(10) == ord('q'):
        break

cv2.destroyAllWindows()
cap.release()
```
