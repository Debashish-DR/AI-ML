```python
#Data Cleaning
import numpy as np
import pandas as pd
```

```python
# delim_whitespace because our columns are separated with spaces
# delimiter also does this work when there is neither comma nor spaces but like -->';' , ':'-->syntax--> delimter=';' 
auto = pd.read_csv(r"D:\jupyter_PY\CTTC-AIML\auto+mpg\auto-mpg.data",header = None,delim_whitespace=True) 
auto
```

```python
#Renaming the columns
auto.columns = ['mpg','cylinder','displacement','horsepower','weight','acceleration','model year','origin','car name']
auto.head()
```

```python
#check missing values
auto.isnull().sum()
```

```python
#Check Datatypes
auto.dtypes
```

```python
#check unique values
for i in auto.columns:
    print(i,':','\n',auto[i].unique(),'\n') #it show the incorrect datatype in horsepower because horsepowe has different dtypes result
```

```python
for i in auto.columns:
    print(i , ':' , sum(auto[i] == '?'))
```

```python
d = auto.describe(include = 'all')
d
```

```python
auto.horsepower.replace('?', d['horsepower'][2],inplace=True) #we take 'top' row because it is the top repeated value
```

```python
# Check once again for null values
for i in auto.columns:
    print(i , ':' , sum(auto[i] == '?'))
```

```python
for i in auto.columns:
    print(i,':','\n',auto[i].unique(),'\n') # now we can see the '?' mark will not be there in horsepower column
```

```python
auto.horsepower = auto.horsepower.astype('float')   #it will not work for -> (inplace = True) ---X
```

```python
auto.dtypes
```

```python
for i in auto.columns:
    print(i,':','\n',auto[i].unique(),'\n') # we can see now that the horsepower column elements has converted to float
```

```python
#Deleting a column
auto.drop('car name',axis=1,inplace=True)
auto.head()
```
